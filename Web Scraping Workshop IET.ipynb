### Request Module

requests.get()=>Downloads files and web pages from the Internet.

```python
import requests
res=requests.get('https://www.amazon.in/s/ref=nb_sb_noss_2?url=search-alias%3Daps&field-keywords=mobiles')
```

```python
res.status_code 
```

```python
res.raise_for_status #Raise an exception if there was an error downloading the file 
                     #and will do nothing if the download succeeded. 
```

```python
type(res) #returns a Response object, which contains the response that the web server gave for our request
```

```python
res.text #downloaded web page is stored as a string in the Response objects text variable
```

```python
myFile = open('Amazonfile.txt', 'wb') #open the file in write binary mode 
                                      #by passing the string 'wb' as the second argument to open()
```

```python
for chunk in res.iter_content(100000):
        myFile.write(chunk)
myFile.close()
```

## Beautiful Soup

Beautiful Soup is a module for extracting information from an HTML page(parse an HTML File).The BeautifulSoup modules name is bs4 (for Beautiful Soup, version 4).

```python
import bs4
beautifulobj=bs4.BeautifulSoup(res.text,"lxml") #The bs4.BeautifulSoup() function returns is a BeautifulSoup object
type(beautifulobj)
```

We can retrieve a web page element from a BeautifulSoup object by calling the select()method and passing a string of a CSS selector for the element you are looking for.

Selectors are like regular expressions: They specify a pattern to look for, in this case, in HTML pages instead of general text strings.

```python
 for loop1 in beautifulobj.findAll('div',{"id":"resultsCol"}):
        for loop2 in loop1.findAll('a'):
            if loop2!=None:
                if loop2.get('href')==None or loop2.get('href')=='javascript:void(0)':
                    continue
                print(loop2.get('href'))
        
```
